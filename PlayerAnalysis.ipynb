```python
def grab_json(req):
    """
    Grab info specified in `spec.json`.
    
    req: str
    Returns: dict
    """
    api = "https://api.opendota.com/api"
    req = api + req
    r = requests.get(req)
    return r.json()

#grab_json("/players/37244929")

Alex = "/players/37244929"
grab_json(Alex + "/recentMatches")
```
